```python
import tkinter as tk
import tkinter.ttk as tt
```

```python
window = tk.Tk()
# 設定視窗標題、大小和背景顏色
window.title('Calculated volume App')
window.geometry('800x600')
window.configure(background='white')
header_label = tk.Label(window, text='Calculated volume')
header_label.pack()
volume_frame=tk.Frame(window)
volume_frame.pack(side=tk.TOP)
volumesize = tk.Label(volume_frame, text = 'Shapes:')
volumesize.pack(side=tk.LEFT)

Shapeskind = ('cube','cylinder','cone','rectangular')

Shapeskindcom = tt.Combobox(volume_frame, width=50, values=Shapeskind)
Shapeskindcom.pack(side=tk.LEFT)
# 以下為 length_frame 群組
length_frame = tk.Frame(window)
# 向上對齊父元件
length_frame.pack(side=tk.TOP)
length_label = tk.Label(length_frame, text='length(cm)/radius(cm)')
length_label.pack(side=tk.LEFT)
length_entry = tk.Entry(length_frame)
length_entry.pack(side=tk.LEFT)

# 以下為 width_frame 群組
width_frame = tk.Frame(window)
width_frame.pack(side=tk.TOP)
width_label = tk.Label(width_frame, text='width(cm)')
width_label.pack(side=tk.LEFT)
width_entry = tk.Entry(width_frame)
width_entry.pack(side=tk.LEFT)

# 以下為 high_frame 群組
high_frame = tk.Frame(window)
high_frame.pack(side=tk.TOP)
high_label = tk.Label(high_frame, text='high(cm)')
high_label.pack(side=tk.LEFT)
high_entry = tk.Entry(high_frame)
high_entry.pack(side=tk.LEFT)

result_label = tk.Label(window)
result_label.pack()

def click_me():
    if Shapeskindcom.get()== 'rectangular':
      length = float(length_entry.get())
      width = float(width_entry.get())
      high = float(high_entry.get())
      volume_value = round(length*width*high, 2)
    
      result_label.configure(text=Shapeskindcom.get()+" 's volume is "+ str(volume_value))
    elif Shapeskindcom.get()== 'cube':
      length = float(length_entry.get())
      width = float(width_entry.get())
      high = float(high_entry.get())
      volume_value = round(length*width*high, 2)
    
      result_label.configure(text=Shapeskindcom.get()+" 's volume is "+ str(volume_value))
    elif Shapeskindcom.get()== 'cylinder':
      length = float(length_entry.get())
      high = float(high_entry.get())
      volume_value = round(length*length*high*3.14, 2)
    
      result_label.configure(text=Shapeskindcom.get()+" 's volume is "+ str(volume_value))
    elif Shapeskindcom.get()== 'cone':
      length = float(length_entry.get())
      high = float(high_entry.get())
      volume_value = round(length*length*high*3.14/3, 2)
    
      result_label.configure(text=Shapeskindcom.get()+" 's volume is "+ str(volume_value))
calculate_btn = tk.Button(window, text='Calculated_volume', command=click_me)
calculate_btn.pack()
# 運行主程式
window.mainloop()
```
